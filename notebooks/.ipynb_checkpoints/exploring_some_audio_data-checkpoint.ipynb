# Computing the periodogram

Using the `scipy.signal` package

Here is the documentation for [computing periodograms.](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.periodogram.html#scipy.signal.periodogram)

### First, get windows

There are two window types referred to in the original pdf: Hanning window and blackman window. Both of these are options in the scipy [windows functions](https://docs.scipy.org/doc/scipy/reference/signal.windows.html#module-scipy.signal.windows) page.

```python
# create a sub sequence of one of our audio channels -- first minute of first column
data_sub = data[:fs, 0]
```

```python
# following example from above -- Hann
window = signal.windows.hann(51)
plt.plot(window)
plt.show()
```

```python
# following example from above -- blackman
window = signal.windows.blackman(51)
plt.plot(window)
plt.show()
```

##### `get_window`

```python
# see what this function outputs
window_hann_10 = signal.windows.get_window(window='hann', Nx=10)
window_hann_20 = signal.windows.get_window(window='hann', Nx=20)

window_blackman = signal.windows.get_window(window='blackman', Nx=10)
window_boxcar = signal.windows.get_window(window='boxcar', Nx=10)
```

```python
# plot the above
plt.figure(figsize=(12.5, 6))

# plot different windows
plt.plot(window_hann_10, label='Hann, $Nx=10$')
plt.plot(window_hann_20, label='Hann, $Nx=20$')
plt.plot(window_blackman, label='Blackman')
plt.plot(window_boxcar, label='Boxcar')

plt.legend(loc='best')
plt.show()
```

### Try getting a periodogram for two different parts of the song

```python
# find another part of the song
data_sub_2 = data[5*fs:6*fs, 0]
```

```python
# plot these two side by side
fig, ax = plt.subplots(2, 1, sharex=True, sharey=True, figsize=(12.5, 6))

ax[0].plot(data_sub)
ax[0].set_title('First second')

ax[1].plot(data_sub_2)
ax[1].set_title('Fifth Second')

plt.show()
```

```python
# frequency versus power
f_1, pxx_den_1 = signal.periodogram(data_sub, fs=fs, window='hann')
f_5, pxx_den_5 = signal.periodogram(data_sub_2, fs=fs, window='hann')
```

```python
# plot these two side by side
fig, ax = plt.subplots(2, 1, sharex=True, sharey=True, figsize=(12.5, 6))

ax[0].semilogy(f_1, pxx_den_1)
ax[0].set_title('First second')

ax[1].semilogy(f_5, pxx_den_5)
ax[1].set_title('Fifth Second')

plt.show()
```

```python
pxx_den_1[np.where(pxx_den_1<20000)]
```

```python
# what happens if we plot this in terms of a histogram?
plt.figure(figsize=(12.5, 6))
plt.hist(np.log(pxx_den_1), bins=100)
plt.show()
```

```python
# what happens if we plot this in terms of a histogram?
plt.figure(figsize=(12.5, 6))
plt.hist(np.log(pxx_den_5), bins=100)
plt.show()
```

### Create a periodogram for another part of the song

```python
# create periodogram for entire song
f_all, pxx_den_all = signal.periodogram(data[:, 0], fs=fs, window='blackman')
```

```python
# plot the above
plt.figure(figsize=(12.5, 6))
plt.plot(f_all, pxx_den_all)
plt.show()
```

##### Sine waves

```python
# number of sample points
N = 600

# sample spacing
T = 1.0 / 800.0
```

```python
# create the data
x = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*x) + np.sin(80.0 * 2.0*np.pi*x)
```

```python
# plot y
plt.figure(figsize=(12.5, 6))
plt.plot(x, y)
plt.show()
```

```python
# apply the fourier transform
yf = fft(y)
xf = np.linspace(0.0, 1.0 / (2.0*T), N//2)
```

```python
# plot yf
plt.figure(figsize=(12.5, 6))
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.show()
```
