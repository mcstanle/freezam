# Computing the periodogram

Using the `scipy.signal` package

Here is the documentation for [computing periodograms.](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.periodogram.html#scipy.signal.periodogram)

### First, get windows

There are two window types referred to in the original pdf: Hanning window and blackman window. Both of these are options in the scipy [windows functions](https://docs.scipy.org/doc/scipy/reference/signal.windows.html#module-scipy.signal.windows) page.

```python
# create a sub sequence of one of our audio channels -- first minute of first column
data_sub = data[:fs, 0]
```

```python
# following example from above -- Hann
window = signal.windows.hann(51)
plt.plot(window)
plt.show()
```

```python
# following example from above -- blackman
window = signal.windows.blackman(51)
plt.plot(window)
plt.show()
```
